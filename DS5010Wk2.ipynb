```python
def readit(filename="iris.csv"):
    with open(filename, 'r') as file:
      data = file.read().rstrip().split("\n")
    
    keys = data.pop(0).split(",")
    list = []
    for line in data:
      datum = {keys[i]: value for i, value in enumerate (line.split(","))}
      list.append(datum)

    for entry in list:
      for key, value in entry.items():    #used looping technique found here: https://docs.python.org/3/tutorial/datastructures.html
        try:
          entry[key] = float(value)
        except:
          if entry[key] == "NA":          #adjusted to filter out NA's in Penguin dataset
            entry[key] = 0
          else:
            pass
          
      
    return list

readit("penguins.csv")

```

```python
import matplotlib.pyplot as plt

flowers = readit("iris.csv")

x = []
y = []
colors = []
for i in range (len(flowers)):
  x.append(flowers[i]['sepal_width'])
  y.append(flowers[i]['sepal_length'])
  if flowers[i]['species'] == 'setosa':
    colors.append('blue')
  elif flowers[i]['species'] == 'versicolor':
    colors.append('orange')
  else:
    colors.append('green')

##Tried to use a matrix to create relationship between species and color, didn't work
# for i in range (len(species)):
#   if species[i] == "setosa":
#     species[i] = ['setosa', 'blue']
#   elif species[i] == 'versicolor':
#     species[i] = ['versicolor', 'orange']
#   else:
#     species[i] = ['virginica', 'green']


##tried to use syntax suggested on Legend guide on matplotlib, failed. https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
#labels = ['setosa', 'versicolor', 'virginica']
# fig, ax = plt.subplots()
# for point in labels:
#   ax.scatter(x, y, c = colors, s = 20, label = labels)

# ax.legend()

plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.title('Iris Scatterplot')
plt.scatter(x, y, 20, c = colors)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

penguins = readit("penguins.csv")

x = []
y = []
colors = []
for i in range (len(penguins)):
  x.append(penguins[i]['flipper_length_mm'])
  y.append(penguins[i]['bill_length_mm'])
  if penguins[i]['species'] == 'adelie':
    colors.append('blue')
  elif penguins[i]['species'] == 'gentoo':
    colors.append('orange')
  else:
    colors.append('green')


plt.xlabel('Flipper Length')
plt.ylabel('Bill Length')
plt.ylim(30, 60)
plt.xlim(150, 250)
plt.title("Penguins Scatterplot")
plt.scatter(x, y, 20, c = colors)

plt.legend()
plt.show()
```
